## Populations data 

```python
population = pd.read_csv('./data/API_SP.POP.TOTL_DS2_en_csv_v2_2445260.csv', skiprows=4)
print(population.shape)
population.head()
```

```python
# we only need 1995-2018
population.drop(useless_cols, axis='columns', inplace=True)

# drop regions
population = population[population['Country Name'].isin(regions) == False]
print(population.shape)
population.head()
```

## Departures per capita

```python
dep_per_capita = pd.merge(outbound, population, on='Country Code', suffixes=('_dep', '_pop'))
print(dep_per_capita.shape)
dep_per_capita.head()
```

```python
for year in range(1995, 2019):
    dep = str(year) + '_dep'
    pop = str(year) + '_pop'
    dep_per_capita[str(year)] = dep_per_capita[dep]/dep_per_capita[pop]
#     dep_per_capita.drop([dep, pop], axis='columns', inplace=True)
dep_per_capita.head()
```

```python
dep_per_capita.drop(['Unnamed: 65', 'Country Name_dep'], axis='columns', inplace = True)
```

```python
dep_per_capita.rename(columns = {'Country Name_pop': 'Country Name'}, inplace=True)
dep_per_capita.head()
```

## Expenditures per departure

```python
exp_per_dep = pd.merge(outbound, expenditures, on='Country Code', suffixes=('_dep', '_exp'))
exp_per_dep.head()
```

```python
for year in range(1995, 2019):
    dep = str(year) + '_dep'
    exp = str(year) + '_exp'
    exp_per_dep[str(year)] = exp_per_dep[dep]/exp_per_dep[exp]
#     exp_per_dep.drop([dep, exp], axis='columns', inplace=True)
exp_per_dep.head()
```

```python
exp_per_dep.drop(['Country Name_dep'], axis='columns', inplace = True)
dep_per_capita.rename(columns = {'Country Name_exp': 'Country Name'}, inplace=True)
dep_per_capita.head()
```

## Country Codes  
amCharts package uses 2-letter codes to identify the countries => need to add these codes to the datasets.  
Data [source](https://www.kaggle.com/juanumusic/countries-iso-codes?select=wikipedia-iso-country-codes.csv)

```python
codes = pd.read_csv('./data/wikipedia-iso-country-codes.csv')
print(codes.shape)
codes.head()
```

```python
codes = codes[['Alpha-2 code', 'Alpha-3 code']]
codes.rename(columns={'Alpha-2 code': 'CountryCode2', 'Alpha-3 code': 'CountryCode'}, inplace=True)
```

```python
dep_per_capita = pd.merge(dep_per_capita, codes, on='CountryCode')
dep_per_capita.head()
```

```python
dep_per_capita.shape
```
